```
# import tensorflow as tf
import pickle
import numpy as np
import math



def load_pickle(filename):
    loaded_content = pickle.load(open("drive/My Drive/Winter 2020/NLP/NLP_project/Project_2/data/"+filename, "rb"))
    return loaded_content


def save_pickle(data, filename):
    pickle.dump(data, open("drive/My Drive/Winter 2020/NLP/NLP_project/Project_2/data/"+filename, "wb"))


def create_batches(corpus, n_batches):
    if len(corpus) % math.ceil(len(corpus)/n_batches) != 0:
      zeros = np.zeros(((math.ceil(len(corpus)/n_batches)) - (len(corpus) % (math.ceil(len(corpus)/n_batches)))))
      corpus = np.append(corpus, zeros, axis=0)
      # batched_corpus = np.reshape(padded_corpus, (math.ceil(len(corpus)/batch_size), batch_size))
    # print(padded_corpus.shape, zeros.shape, corpus.shape, batch_size)
    # else:
    batched_corpus = np.reshape(corpus, (n_batches, math.ceil(len(corpus)/n_batches)))
    print(batched_corpus.shape)
    return batched_corpus



train, test, valid = load_pickle("training_step3.pkl"), load_pickle("testing_step3.pkl"), load_pickle("valid_step3.pkl")
n_batches = 30
train = create_batches(np.array(train), n_batches)
sliding_window_size = 5
save_pickle(train[0][:sliding_window_size], "training_sequence1.pkl")
save_pickle(train[1][sliding_window_size:2*sliding_window_size], "training_sequence2.pkl")
print('done')

```
